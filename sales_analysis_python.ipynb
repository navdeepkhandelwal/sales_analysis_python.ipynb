```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (8,5)
```

```python
df = pd.read_excel("Superstore.xls")
df.head()
```

# New Section

```python
df.shape
df.columns
df.info()
df.describe()
```

```python
# Missing values
df.isnull().sum()
df.dropna(inplace=True)

# Convert dates
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Remove duplicates
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
```

```python
df['Sales'].sum()
df['Profit'].sum()
```

```python
df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('Region')['Sales'].sum()
```

```python
df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Category")
plt.show()
```

```python
df.groupby('Month')['Sales'].sum().plot()
plt.title("Monthly Sales Trend")
plt.show()
```

```python
df.groupby('Region')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title("Region-wise Sales")
plt.show()
```

```python
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Profit vs Sales")
plt.show()
```

Technology category generates highest sales

1. Technology category generates highest sales  
2. Some regions have higher sales contribution
3. High sales doesn’t always mean high profit
4. Discounts may reduce profitability
